```python
#dropping the last column as its empty
data_cleaned= data.drop(['Unnamed: 32'], axis=1)
print(f"shape after dropping the last column: {data_cleaned.shape}")
```

```python
#handling outliers by removing them via IQR method
def remove_outliers(df):
  Q1 = df.quantile(0.25)
  Q3 = df.quantile(0.75)
  IQR = Q3 - Q1
  return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

data_no_outliers = remove_outliers(data_cleaned.drop('diagnosis',axis=1))

#adding 'diagnosis' back to cleaned data
data_no_outliers['diagnosis'] = data_cleaned['diagnosis']

print(f"Shape after removing outliers: {data_no_outliers.shape}")
```
